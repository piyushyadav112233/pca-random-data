```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Step 1: Generate or Load the Dataset
#random data
np.random.seed(0)
n_samples = 1000
age = np.random.randint(18, 65, n_samples)
income = np.random.randint(20000, 100000, n_samples)
spending = np.random.randint(10, 50, n_samples)
df = pd.DataFrame({'Age': age, 'Income': income, 'Spending': spending})
print(df.head())

```

```python
# Step 2: Standardize the Data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
print(df_scaled)
```

```python
# Step 3: Apply PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)
print(df_pca)
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
print(df_pca)
```

```python
# Step 4: Create a DataFrame with PCA Results
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
print(df_pca)
```

```python
#visulazation  step 5
plt.figure(figsize=(10, 6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization')
plt.grid(True)
plt.show()
```

```python
# Step 6: Output Explained Variance
explained_variance = pca.explained_variance_ratio_
print(explained_variance)
```

```python
# Step 7: Save the Dataset and PCA Results
original_file_path = 'original_data.csv'  # Assign the filename to original_file_path
pca_file_path = 'pca_results.csv'  # Assign the filename to pca_file_path
df_pca.to_csv(pca_file_path, index=False)
df.to_csv(original_file_path, index=False)
print("Data saved successfully.","PCA results saved successfully.")

print(f"Original dataset saved to {original_file_path}")
print(f"PCA results saved to {pca_file_path}")
```
